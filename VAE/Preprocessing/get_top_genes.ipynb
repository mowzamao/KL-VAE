```python
import pandas as pd
import numpy as np

def get_top_genes(num_genes, by):
    if by == 'kl':
        fname = "/home/sophiemowe/cits4010/cits4010/VAE/VAE/data/documentation/top_genes_by_kl.csv"
    if by == 'frequency':
        fname = "/home/sophiemowe/cits4010/cits4010/VAE/VAE/data/documentation/top_genes_by_frequency.csv"
    else:
        fname = '/home/sophiemowe/cits4010/cits4010/VAE/VAE/data/documentation/random_genes.csv'

    top_genes = pd.read_csv(fname)
    if num_genes == 0:
        return top_genes
    return top_genes.head(num_genes-1)

def get_top_gene_indices(num_genes, by):
    top_genes = get_top_genes(num_genes, by)
    top_genes = top_genes['index'].values
    top_genes = np.append(top_genes, 60660)        
    return top_genes

def save_top_genes_dataset(dataset, dataname, num_genes, top_gene_indices, by):
    fname = f'/home/sophiemowe/cits4010/cits4010/VAE/VAE/data/top_genes/by_{by}/{dataname}'
    dataset = dataset[:, top_gene_indices]
    print(f'{dataname} shape: {dataset.shape}')
    print(dataset[0])
    np.save(f'{fname}_top_{num_genes}_genes_{by}.npy',dataset)
    print("Saved dataset with top genes to ", fname)
    return dataset
```

```python
num_genes = 1000
for by in ['frequency', 'kl', 'random']:
    print(f"Getting top {num_genes} genes")
    top_gene_indices = get_top_gene_indices(num_genes, by)

    train_scaled = np.load('/home/sophiemowe/cits4010/cits4010/VAE/VAE/data/full_dataset/train_scaled.npy')
    val_scaled = np.load('/home/sophiemowe/cits4010/cits4010/VAE/VAE/data/full_dataset/val_scaled.npy')

    save_top_genes_dataset(train_scaled, 'train', num_genes, top_gene_indices, by)
    save_top_genes_dataset(val_scaled, 'val', num_genes, top_gene_indices, by)
```
